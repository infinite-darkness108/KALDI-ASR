```python
data = pd.read_csv('mfcc40.csv')
```

```python
data = data.dropna()
data = data.drop(['participant','audio id'], axis = 1)
x = data.drop(['severity'], axis=1)
y = data['severity']
y
```

```python
from sklearn.preprocessing import LabelEncoder
label_en=LabelEncoder()
y=label_en.fit_transform(y)
y
```

```python
for i in x.columns:
  if x[i].dtype == object:
    x[i] = label_en.fit_transform(x[i])
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
# Random Forest

from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=32)
rf_classifier.fit(x_train,y_train)
y_pred = rf_classifier.predict(x_test) 
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

print("Accuracy is: ", accuracy_score(y_test,y_pred))
print("Precision is: ", precision_score(y_test,y_pred, average = 'weighted'))
print("Recall is: ", recall_score(y_test,y_pred, average = 'weighted'))
print("F1 score is: ", f1_score(y_test,y_pred, average = 'weighted'))
```

```python
from sklearn import svm, datasets
import sklearn.model_selection as model_selection
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
rbf = svm.SVC(kernel='rbf', gamma=5, C=25).fit(x_train, y_train)
poly = svm.SVC(kernel='poly', degree=4, C=2).fit(x_train, y_train)
lin = svm.SVC(kernel='linear', degree=4, C=25).fit(x_train, y_train)
```

```python
rbf_pred = rbf.predict(x_test)
poly_pred = poly.predict(x_test)
lin_pred = lin.predict(x_test)
poly_accuracy = accuracy_score(y_test, poly_pred)
poly_f1 = f1_score(y_test, poly_pred, average='weighted')
print('Accuracy (Polynomial Kernel): ', "%.2f" % (poly_accuracy*100))
print('F1 (Polynomial Kernel): ', "%.2f" % (poly_f1*100))
rbf_accuracy = accuracy_score(y_test, rbf_pred)
rbf_f1 = f1_score(y_test, rbf_pred, average='weighted')
print('Accuracy (rbf Kernel): ', "%.2f" % (rbf_accuracy*100))
print('F1 (rbf Kernel): ', "%.2f" % (rbf_f1*100))
lin_accuracy = accuracy_score(y_test, lin_pred)
lin_f1 = f1_score(y_test, lin_pred, average='weighted')
print('Accuracy (linear Kernel): ', "%.2f" % (lin_accuracy*100))
print('F1 (linear Kernel): ', "%.2f" % (lin_f1*100))
```

```python
d = []

path_audios = glob.glob(os.path.abspath('/home/sltlab/anaconda3/envs/speech/TDSC/*/*/audio/*.wav'))
#print(path_audios)
for i in path_audios:
    #print("hi")
    c = i.split('/')
    y, sr = librosa.load(i)
    #mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40)
    mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=40)
    log_mel_spectrogram = librosa.power_to_db(mel_spectrogram, ref=np.max)
    task = c[-4]
    participant = c[-3]
    a = c[-1]
    data = {
        'severity': task,
        'participant': participant,
        'audio id':a
    }
    #print(log_mel_spectrogram)
    #df = f.reset_index(drop=True)
    #p = mfcc.to_dict('records')
    #print(p)
    for i in range(0,len(log_mel_spectrogram)):
        try:
            data.update({"log_mel_energy_"+str(i):log_mel_spectrogram[0][i]})
        except IndexError:
            pass

    #data.update(p)
    d.append(data)
```
